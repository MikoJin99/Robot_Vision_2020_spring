## read images

#### The rgb image file is "taken" from https://www.druckdiscount24.de/faq/farben

```python
%matplotlib inline
# %matplotlib qt
imgfile = './data/Farbkreis-RGB.jpg'
imgfile_Pepper = './data/Peppers.tiff'
im_cv = cv2.imread(imgfile)
im_Pepper = cv2.imread(imgfile_Pepper)

converted_frame0= cv2.cvtColor(im_cv, cv2.COLOR_BGR2RGB) # convrt color bgr to rgb
fig= plt.figure(figsize=(10,10))
plt.imshow(converted_frame0)
```

```python
%matplotlib inline
# %matplotlib qt

fig = plt.figure(figsize = (8,8))

ax1 = fig.add_subplot(1, 3, 1, frameon = False)  # Red
ax1.imshow(converted_frame0[:,:,0], cmap = 'gray')
plt.title("Red")

ax2 = fig.add_subplot(132, frameon=False) # Green
ax2.imshow(converted_frame0[:,:,1], cmap = 'gray')
plt.title("Green")

ax3 = fig.add_subplot(133, frameon=False) # Blue
ax3.imshow(converted_frame0[:,:,2], cmap = 'gray')
plt.title("Blue")

plt.show()
```

## change RGB to HSV

```python
HSV = cv2.cvtColor(converted_frame0, cv2.COLOR_RGB2HSV_FULL) # convrt color bgr to rgb
```

```python
import matplotlib.pylab as plt
%matplotlib inline
# %matplotlib qt

fig = plt.figure(figsize = (8,8))

ax1 = fig.add_subplot(1,4,1, frameon = False)  # Red
ax1.imshow(converted_frame0)
plt.title("RGB")


ax2 = fig.add_subplot(1,4,2, frameon = False)  # Red
ax2.imshow(HSV[:,:,0], cmap = 'gray')
plt.title("Hue")

ax3 = fig.add_subplot(143, frameon=False) # Green
ax3.imshow(HSV[:,:,1], cmap = 'gray')
plt.title("Saturation")

ax4 = fig.add_subplot(144, frameon=False) # Blue
ax4.imshow(HSV[:,:,2], cmap = 'gray')
plt.title("Value")
```
